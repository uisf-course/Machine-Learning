```python
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(7, 7))

fig.tight_layout()

# plt.style.use("grayscale")

x = np.arange(-6, 7, 1)

y1 = np.power(x, 1)
y2 = np.power(x, 2)
y3 = np.power(x, 3)
y4 = np.power(x, 4)

_ = axes[0, 0].plot(x, y1)
_ = axes[0, 0].set_title("$f(x) = x$")
# _ = axes[0, 0].set_xticks([])

_ = axes[0, 1].plot(x, y2)
_ = axes[0, 1].set_title("$f(x) = x^2$")

_ = axes[1, 0].plot(x, y3)
_ = axes[1, 0].set_title("$f(x) = x^3$")

_ = axes[1, 1].plot(x, y4)
_ = axes[1, 1].set_title("$f(x) = x^4$")
```

```python
fig, ax = plt.subplots(figsize=(4, 4))

y = np.random.randn((9000))

_ = ax.hist(y, bins=50)
_ = ax.set_title("normal distribution")
_ = ax.set_xlabel("x")
_ = ax.set_ylabel("y")
```
